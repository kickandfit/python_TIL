```python
#2447
def stars(n):
    matrix=[]
    for i in range(3 * len(n)):
        if i // len(n) == 1:
            matrix.append(n[i % len(n)] + " " * len(n) + n[i % len(n)])
            print(f'{i}번째 matrix[{i}]에 {matrix[i]}추가했다!!!!')
        else:
            matrix.append(n[i % len(n)] * 3)
            print(f'{i}번째 matrix[{i}]에 {matrix[i]}추가했다')
    return(list(matrix))

star = ["***","* *","***"]
n = int(input())
k = 0
while n != 3:
    n = int(n / 3)
    k += 1
    
for i in range(k):
    star = stars(star)
    print(f'{i+1}번 반복했다')
for i in star:
    print(i)
```
