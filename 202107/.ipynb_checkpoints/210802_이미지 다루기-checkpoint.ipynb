### 이미지 흐리게 만들기
- 영상의 특정 좌표 값을 주변 픽셀 값들의 산술 평균으로 설정하는 방법
- 
- cv2.blur(src, ksize, dst=None, anchor=None, borderType=None) -> dst
 * src: 입력 영상
 * ksize: 평균값 필터 크기. (width, height) 형태의 튜플.
 * dst: 결과 영상. 입력 영상과 같은 크기 & 같은 타입.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread('c:/pydata/plane.jpg', cv2.IMREAD_GRAYSCALE)

# 흐리게 만들기
img_blu=cv2.blur(img, (5, 5))   # 주변의 픽셀을 5x5 커널 평균값으로 이미지 흐리기
plt.imshow(img_blu, cmap="gray")
plt.show()
```

```python
# 흐리게 만들기
img_blu=cv2.blur(img, (100, 100))
plt.imshow(img_blu, cmap="gray")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def bluring():
    img=cv2.imread('c:/pydata/plane.jpg')
    
    kernel=np.ones((5,5), np.float32)/25
    blur=cv2.filter2D(img, -1, kernel)
    
    cv2.imshow('original', img)
    cv2.imshow('blur', blur)
    
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
bluring()
```

### 이미지에 대한 색상 구분하기

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_bgr=cv2.imread('c:/pydata/plane_1.jpg')

# BGR에서 HSV로 변환 : https://www.charlezz.com/?p=44906
img_hsv=cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)

# HSV에서 파랑 값 범위 정의
lower_blue=np.array([50, 100, 50])
upper_blue=np.array([130, 255, 255])

# 마스크 생성
mark=cv2.inRange(img_hsv, lower_blue, upper_blue)

#mark 적용
img_hsv_mark=cv2.bitwise_and(img_bgr, img_bgr, mark)

# plt에서 출력하기 위해 RGB 형식으로 변경
img_rgb=cv2.cvtColor(img_hsv_mark, cv2.COLOR_HSV2RGB)

# 결과확인
plt.imshow(img_hsv_mark, cmap="gray")
plt.show()

plt.imshow(mark, cmap="gray")
plt.show()
```

### 배경 제거
- https://docs.opencv.org/3.4/d8/d83/tutorial_py_grabcut.html

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_bgr=cv2.imread('c:/pydata/plane_1.jpg')
mask = np.zeros(img_bgr.shape[:2],np.uint8)
print(mask.shape)

bgdmod = np.zeros((1,65),np.float64)
fgdmod = np.zeros((1,65),np.float64)

rect = (0, 50,250,110)
cv2.grabCut(img_bgr, mask, rect, bgdmod, fgdmod, 5, cv2.GC_INIT_WITH_RECT)



mask2 = np.where((mask==0)|(mask==2),0,1).astype('uint8')  
img_rgb = img_bgr*mask2[:,:,np.newaxis]

plt.imshow(img_rgb)
plt.show()
```
