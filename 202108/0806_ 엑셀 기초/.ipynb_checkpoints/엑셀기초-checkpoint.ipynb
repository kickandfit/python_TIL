# 자료 타입

- 문자 데이터

# 수식에 대해 알아보기

- 수식 입력시 : = 입력을 시작으로 한다
    - = 2+1 --------> 3 출력(B3 이라고 가정)
    - 셀참조를 활용할 수 있음
        - = B3 + 1(수식입력줄에 보이는 형태) ------->4 
        - = B3 + B5 -----> 7
        - 문자 데이터는 산술 연산 불가능 (#VALUE! 가 나타남)
        - = B4 & B5 ( & 문자열 연결 연산 : 두개의 데이터를 연결하여 문자로 만들어줌) -------> 34(문자열)
        - ="50" (등호로 시작하고 "" 안에 입력하면 문자 데이터로 인식)
        - 특이한 것은 50 은 문자지만 숫자와 산술연산이 가능하고 숫자로 인식
        - 맨처음 = 수식임을 나타내고, 다음 등호는 같다 라는 것을 나타냄
        - ="50"=50-----------> False ( 앞의 것 문자 뒤의 것 숫자)
        - ="50"*1 = 50 ---->True
      
- 연산의 종류 ( 순서대로 우선순위 )

 - 참조연산
 - 산술연산
 - 문자열연산
 - 비교연산  : T/F ---->비교대상/기준값을 넣음( 결과는 T/F )
 - 함수식 연산

# 셀참조

- 상대참조
    - B5 와 같은 것
        - 채우기 핸들 또는 셀을 복사해서 특정 위치에 붙여넣기하면 이동 만큼 참조위치가 변함
    
- 절대참조(F4 눌러서 조정)
    - '$달러B$달러5'
        - 채우기 핸들 또는 셀을 복사해서 특정 위치에 붙여넣기 해도 항상 동일한 셀/범위를 참조
        - 시트이름!셀 (참조할 셀이 다른 시트인경우)
           (ex > 기본자료!달러c달러3)
        - 다른 셀에 이름상자에 이름 정의하면 수식에서 사용가능
        - = 셀 * F3(내가 저장한 셀 이름들 나옴/ 선택)
    
- 혼합참조
    - '$달러B5,B$달러5'
        - 셀 참조시 행 또는 열 하나는 상대 다른 하나는 절대 참조 형식
        - 마우스로 드래그하면 서식이 깨짐
        - 따라서 동일한 수식을 적용하는 방법은 범위를 선택 적용 수식 셀 선택 후 ctrl + Enter

# 찾기 참조 함수

- HLOOKUP(찾을값, 대상범위, 출력행위치, 찾는 방법)
- VLOOKUP(찾을값, 대상범위, 출력열위치, 찾는 방법)
    - 대상 범위가 있어야 함
    - HLOOKUP은 행단위로 들어가 있어야함
        - 행에 INDEX가 있는 경우 사용
        - 찾아야하는 대상이 범위의 첫번째 행이 있다면 그 대상 기준 열이 출력대상이고 출력 행의 위치에 해당하는 값을 출력)
        - 첫번째 열에서 먼저 찾고 그다음 행 위치
    - VLOOKUP은 열단위로 들어가 있어야함
        - 한국에서는 VLOOKUP이 많이 쓰임
            - N/A 에러 : 찾는 값이 없다 ( 마지막 숫자가 0 인경우)
            - 1 은 유사 일치값( 범위값/찾을값과 같거나 작은 가장 가까운 값)
            - 0 은 고유값(반드시 존재하는 값)
            - 1을 쓴다면 재미있는 상황이 발생하는데, 다음 행/열을 비교할 때 자신 보다 큰 값이 나온 이후에 바로전 값을 선택
