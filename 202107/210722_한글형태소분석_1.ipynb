```python
from konlpy.tag import Okt
from nltk import Text
from wordcloud import WordCloud
```

```python
f=open("c:/pydata/2021_신년사.txt", "r")
texts=f.read()
f.close()
```

```python
print(texts[:50])
```

```python
t=Okt()
print(t.tagset)
```

```python
pos_txt=t.pos(texts)    # 단어(토픽)을 품사 단위로 출력
print(pos_txt)
```

```python
nouns_txt=t.nouns(texts)    # 토픽들 중에서 명사만 출력
print(nouns_txt)
```

```python
a=1; b=2
aa="a의 값은 " + str(a) +",\nb의 값은 " + str(b) + "입니다."
print(aa)

aa=f'a의 값은 {a}, \n b의 값은 {b}입니다.'
print(aa)
```

```python
### 형용사만 추출
형용사=[t[0] for t in pos_txt if t[1]=='Adjective']
명사=[t[0] for t in pos_txt if t[1]=='Noun']
동사=[t[0] for t in pos_txt if t[1]=='Verb']

#print(f'[형용사]\n{형용사}\n')
#print(f'[명사]\n{명사}\n')
#print(f'[동사]\n{동사}\n')
```

```python
형용사=[];명사=[];동사=[]

for t in pos_txt:
    if t[1]=="Adjective":
        형용사.append(t[0])
    elif t[1]=="Noun":
        명사.append(t[0])
    elif t[1]=="Verb":
        동사.append(t[0])

#print(f'[형용사]\n{형용사}\n')
#print(f'[명사]\n{명사}\n')
#print(f'[동사]\n{동사}\n')
```

```python
morph=명사+형용사+동사
#print(len(morph), "\n", morph)
```

```python
from collections import Counter   # JDK에서 제공하는 기본 라이브러리
```

```python
morph_cnt=Counter(morph)    # 톡픽(단어)에 대한 빈도수를 딕셔너리 구조로 제공
#print(morph_cnt)
morph_txt=morph_cnt.most_common()    # 딕셔너리 구조의 데이터를 내림차순 정렬 후 튜플 구조로 변경
#print(f'[전  체]\n{morph_txt}')
```

```python
### 현재 나타난 토픽에서 5회 이상 거론된 토픽만 추출
     
for i in range(len(morph_txt)):
    for a,b in morph_txt:
        if b < 5:
            morph_txt.remove((a,b)) # list.remove()  ?? => 각자 찾아보기

print(f'[수정후]\n{morph_txt}')
```

```python
morph_txt=dict(morph_txt)
print(morph_txt)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wordcloud = WordCloud(font_path="c:/Windows/Fonts/malgun.ttf",
                      background_color="white", width=1000, height=1000,
                      max_words=100, max_font_size=300)

wc = wordcloud.generate_from_frequencies(morph_txt)
plt.figure(figsize=(8,15))
plt.imshow(wc)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
