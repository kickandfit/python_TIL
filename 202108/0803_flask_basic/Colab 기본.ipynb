```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_bgr=cv2.imread('./data/plane.jpg')
image_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show()

mask = np.zeros(img_bgr.shape[:2],np.uint8)

bgdmod = None #np.zeros((1,65),np.float64)  # 배경을 체크할 메모리 준비
fgdmod = None #np.zeros((1,65),np.float64)  # 전경을 체크할 메모리 준비

rect = (900, 500, 2600, 1350)  # 전경 크기 설정
cv2.grabCut(img_bgr, mask, rect, bgdmod, fgdmod, 5, cv2.GC_INIT_WITH_RECT)

mask2 = np.where((mask==0)|(mask==2),0,1).astype('uint8')  
img_rgb = img_bgr*mask2[:,:,np.newaxis]

plt.imshow(img_rgb)
plt.show()

plt.imshow(mask)
plt.show()

plt.imshow(mask2)
plt.show()
```
