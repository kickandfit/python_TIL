### ROC CURVE
- 참고문서: https://angeloyeo.github.io/2020/08/05/ROC.html, https://moons08.github.io/datascience/classification_score_roc_auc/   

```python
# [0] : 정상 햄버거, [1] : 유통기한이 지난 햄버거
y_true = [0, 1, 1, 1, 1, 0, 1, 1, 0, 0]

# 각 클래스에 대한 확률은 임의로 지정합니다.
A_proba = [0.6, 0.7, 0.7, 0.8, 0.9, 0.7, 0.85, 0.7, 0.65, 0.75]
B_proba = [0.05, 0.05, 0.1, 0.3, 0.6, 0.3, 0.4, 0.5, 0.2, 0.1]

# 완벽한 모델
C_proba = [0, 1, 1, 1, 1, 0, 1, 1, 0, 0]
```

image.png

```python
from sklearn.metrics import roc_curve, auc

# FPR: 가짜 양성 비율
# TPR: 진짜 양성 비율
fpr_A, tpr_A, thr_A = roc_curve(y_true, A_proba)
fpr_B, tpr_B, thr_B = roc_curve(y_true, B_proba)
fpr_C, tpr_C, thr_C = roc_curve(y_true, C_proba)
```

```python
print('fpr_A, tpr_A, thr_A: ', fpr_A, tpr_A, thr_A)
print('fpr_B, tpr_B, thr_B: ', fpr_B, tpr_B, thr_B)
print('fpr_C, tpr_C, thr_C: ', fpr_C, tpr_C, thr_C)
```

```python
auc_A = auc(fpr_A, tpr_A)
auc_B = auc(fpr_B, tpr_B)
auc_C = auc(fpr_C, tpr_C)
print('auc_A:', auc_A,
      '\nauc_B:',auc_B,
      '\nauc_C:',auc_C)
```

```python
import matplotlib.pyplot as plt

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve', fontsize = 20)

plt.plot(fpr_A, tpr_A, color='darkorange',
         lw=2, label='ROC curve A (area = %0.2f)' % auc_A)
plt.plot(fpr_B, tpr_B, color='blue',
         lw=2, label='ROC curve B (area = %0.2f)' % auc_B)
plt.plot(fpr_C, tpr_C, color='green',
         lw=2, label='ROC curve C (area = %0.2f)' % auc_C)

plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')
plt.legend(loc="lower right")
plt.show()
```
