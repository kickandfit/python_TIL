```python
import pandas as pd

#윤성님 크롤링 자료 읽어오기
df=pd.read_csv('./test.csv')
df2=df.set_index('요리')
#이미지 인식한 재료 
재료=['양파','간장']
# dataframe 에서 음식 관련 딕셔너리 만들기
음식=dict()
#재료 정규식 표현으로 바꾸기
re_재료=[]
for i in range(len(df['재료'])):
    b=df['재료'][i].replace("'",'').replace('[',"").replace(']',"").replace('\n','').replace('\t','').replace(' ','')
    re_재료.append(b.split(','))
재료리스트=[]
for i in re_재료:
    a=(str(i).replace('(중)','').replace('감자(중간사이즈)','감자').replace('감자큰거','감자').replace('다진','').replace('통','')
                    .replace('노랑','').replace('빨강','').replace('계란삶을때들어가는식초','식초').replace('고추(or부추)것','').replace('김치(익은것)','김치')
                    .replace('다시마5cmx5cm','다시마').replace('다시물(맹물도ok)','다시물').replace('대파10센티','대파').replace('대파것','대파')
                    .replace('초록','').replace('노란','').replace('송송썬',"").replace('부분','').replace('센티','').replace('(3x3)',"")
                    .replace('혹은',"").replace('또는','').replace('사이즈','').replace('전복(특대)','전복').replace('잔파(쪽파)','쪽파')
                    .replace('큰',"").replace('빨간','').replace('(1팩)','')).replace('작은',"").replace('것','').replace('20','').replace('앞다리살','').replace('볶음용','').replace('색',"")   
    a=a.replace("'",'').replace('[',"").replace(']',"").replace('\t','').replace(' ','')
    재료리스트.append(a.split(','))
    
df['재료']=재료리스트
 
#음식 딕셔너리에 요리 이름을 key값으로 재료를 value 값으로 넣기
for i in range(len(df['요리'])):
    음식[df['요리'][i]]=df['재료'][i]
#3개 뽑기 위한 cnt
cnt=0
#겹치는 재료 숫자를 a리스트에 넣기
a=[]
for i in 음식:
    a.append(len(set(음식[i])&set(재료)))
#추천리스트= 음식 순서대로 3가지, 추가재료가 적은 리스트 아직 미구현
추천리스트=[]
추가재료가_적은_리스트=[]
asdf=[]
zzz = []
정렬추천리스트=[]
#max(a)는 겹치는 재료의 최대값
for i in 음식:
    if max(a)==0:
        print('맞는 재료가 없습니다')
        break
    if len(set(음식[i])&set(재료))==len(음식[i]):
        print('있는 재료로 가능한음식 '+'\n'+i+'\n'+df2.loc[i]['레시피'])
        추천리스트.append(i)
    elif len(set(음식[i])&set(재료))==max(a):
        추천리스트.append(i)
        asdf.append(len(set(음식[i])-set(재료)))
        zzz.append({i:len(set(음식[i])-set(재료))})
asdf=list(set(sorted(asdf)))
# 요리의 이름을 인덱스로 넣고 레시피 찾기
for i in asdf :
    for s in 추천리스트 :
        if i ==len(set(음식[s])-set(재료)) :
            정렬추천리스트.append(s)
정렬추천리스트=정렬추천리스트[:3] 
df2=df.set_index('요리')
if len(추천리스트)>3:
    for i in 추천리스트:
            print(i+'\n부족한재료 '+str(set(음식[i])-set(재료)))
            print(df2.loc[i]['레시피'])
            cnt+=1
            if cnt==3:
                break
for i in 정렬추천리스트:
    
    print(i+'\n부족한재료 '+str(set(음식[i])-set(재료)))
    print(df2.loc[i]['레시피'])
```
