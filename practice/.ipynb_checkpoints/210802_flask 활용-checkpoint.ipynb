```python
def move(x,y, i):
    if -1>=x>10 or -1>=y>10:
        return
    print(x,y)
    if i == 0:
        return move(x+1,y , 0)
    elif i == 1:
        return move(x, y+1, 1)
    elif i == 2:
        return move(x-1, y, 2)
    elif i == 3:
        return move(x, y-1, 3)
    else:
        return move(x+1, y-1, 4)

for dir in range(5):
    move(4,4,dir)
```
