# 파이썬 모듈과 객체탐색
- REPL 과 많은 시간을 보내라

```python
#예제 1
import datetime
print(dir(datetime))
```

#### dir()
- 파이썬 REPL에서 클래스에서 찾을 수 있는 메서드와 속성을 알아볼 수 있음
- 모듈을 검사
    - 클래스를 검사해라
- 속성 리스트에서 흥미 있는 것을 추려낼 수 있는 기법
    - [_ for _ in dir(datetime) if 'date' in _.lower()]

```python
# 클래스 검사하기
print(dir(datetime.date))
```

```python
# date 라는 낱말이 포함된 이름만 출력하도록 dir(datetime) 호출 결과 필터링

[_ for _ in dir(datetime) if 'date' in _.lower()]# lower()을 통해 대소문자를 구별하지 않도록 했음
```

#### 객체에서 속성 목록을 가져오는 것으로 문제가 해결되지 않는다면
- help() 라는 내장함수를 사용해라

```python
help(datetime)
```

#### keypoint( 파이썬 객체와 모듈 탐색 )
- 내장된 dir() 함수를 사용하여 인터프리터 세션에서 파이썬 모듈과 클래스를 탐색가능
- help()가 내장되어 있어 인터프리터에서 문서로 찾아볼 수 있음 ( 종료 : q )

#  바이코드로 내부 엿보기

- CPython 인터프리터는 프로그램을 실행할 때 먼저 프로그램을 바이트코드 명령어 시퀀스로 변환
- 바이트코드는 파이썬 가상 시스템이 사용하는 중간 언어, 성능 최적화를 위해 사용
- 사람이 읽을 수 있는 소스 코드를 직접 실행하는 대신 컴파일러 구문 및 의미 분석 결과를 나타내는 숫자 코드, 상수, 참조가 사용됨

#### keypoint( 바이트코드로 내부 엿보기 )
- CPython은 먼저 프로그램을 중간 바이트코드로 변환한 다음 스택 기반 가상 머신에서 바이트코드 실행
- 내장 dis 모듈을 사용하여 내부를 들여다보고 바이트코드를 검사할 수 있음
- 가상 머신을 연구해라 그만한 가치가 있음
