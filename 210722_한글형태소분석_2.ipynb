```python
from konlpy.tag import Okt
from nltk import Text
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt
import nltk

```

```python
f=open("c:/pydata/2021_신년사.txt", "r")
texts=f.read()
f.close()
```

```python
### 형용사/명사/동사만 추출
t=Okt()
pos_txt=t.pos(texts)

형용사=[t[0] for t in pos_txt if t[1]=='Adjective']
명사=[t[0] for t in pos_txt if t[1]=='Noun']
동사=[t[0] for t in pos_txt if t[1]=='Verb']
```

```python
morph=명사+형용사+동사
#print(morph)
morph=Text(morph)
print(morph)
```

```python
plt.title('2021년 신년사 톡픽 빈도수')
morph.plot()        # 각 단어(토큰)의 사용 빈도를 그래프로 표현(Text 클래스가 제공)
plt.show()
```
