```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
from urllib.parse import quote, unquote
from selenium.webdriver.common.keys import Keys
import urllib.request
import os
from tqdm import tqdm
import time
import re 
import pandas as pd
```

```python
def news_parser(Keyword):
    url = 'https://www.joongang.co.kr/search/news?keyword='+Keyword

    driver = webdriver.Chrome('c:/pydata/chromedriver.exe')
    time.sleep(2)
    driver.get(url)

    while True:
        try:
            driver.find_element_by_css_selector("#container > section > div > section > div > a").click()
            time.sleep(2)

            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)

        except:
            break    

    html_source=driver.page_source
    soup=bs(html_source, 'html.parser')
    news_soup=soup.find('ul', class_="story_list")
    
    li_soup=news_soup.find_all('li')
    
    news_list=[]
                    
    for li in li_soup:
        title=li.find('h2', class_="headline").text
        a_url=li.find('a')['href']
        body=li.find('p', class_="description sm_hidden").text
        news_list.append({'제목':title, '내용':body, '상세url':a_url})

    df=pd.DataFrame(news_list, columns=['제목', '내용', '상세url'])
    
    return df   
```

```python
df_test = news_parser('메가존')
```

```python
# 하위 url을 이용한 상세내용 가져오기
title_list=[]
body_list=[]

for df_url in df_test['상세url']:
    # 하위 url 파씽
    html=urlopen(df_url)
    soup=bs(html, 'html.parser')
    
    # 원하는 데이터 추출
    title=soup.find('h1', class_="headline").text

    try:
        body=soup.find('div', class_="article_body fs3").text
    except:
        body=''
        
    #데이터 정규화
    
    title_list.append(re.sub("[^ 0-9ㄱ-ㅣ가-힣a-zA-Z]", r'',title)) #숫자, 한글, 영문을 제외하고 모두 '' 로 변경)
    body_list.append(re.sub("[^ 0-9ㄱ-ㅣ가-힣a-zA-Z]",  r'',body)) #숫자, 한글, 영문을을 제외하고 모두 '' 로 변경)
    
df2=pd.DataFrame({'제목':title_list, '내용':body_list})
```
